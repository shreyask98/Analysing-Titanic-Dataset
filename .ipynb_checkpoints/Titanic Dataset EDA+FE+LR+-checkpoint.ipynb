```python
def analysis(col):
    count = train.groupby(col)["Survived"].value_counts(normalize=True)
    count_df = pd.DataFrame(count)
    count_df.rename(columns = {"Survived" : "percentage"}, inplace = True)
    count_df.reset_index(inplace = True)
    count_df.sort_values(by = ["Survived", col], inplace=True)
    plt.figure(figsize = (10, 10))
    index = 1
    for c in list(count_df[col].unique()):
        plt.subplot(1,len(list(count_df[col].unique())),index)
        count_df.groupby(col)["percentage"].get_group(c).plot.pie(subplots=True,
                                                 labels=list(count_df['Survived'].unique()),
                                                 autopct="%.2f%%",
                                                 textprops={'fontsize': 10},
                                                 )
        plt.title(c, fontdict = {'fontsize': 14})
        index += 1
```

```python
for col in categorical_features:
    analysis(col)
```

## Feature Engineering
### Removal of outliers

```python
train1 = train.copy()
def remove_outliers_IQR(col):
    # Finding the IQR
    percentile25 = train[col].quantile(0.25)
    percentile75 = train[col].quantile(0.75)
    iqr = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    train[col] = np.where(train[col]>upper_limit, upper_limit, np.where(train[col]<lower_limit,lower_limit,train[col]))
    return train[train[col] > upper_limit]
```

```python
for col in numeric_features:
    remove_outliers_IQR(col)
```
